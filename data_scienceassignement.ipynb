```python
import ast
import re
import datetime

# Restore 'info.dates' to its original string state from the CSV, if it was altered by previous erroneous runs
df['info.dates'] = pd.read_csv('/content/International_T20_Data.csv')['info.dates']

def parse_date_entry_hybrid(date_str_entry):
    if not isinstance(date_str_entry, str):
        return pd.NaT

    # Attempt 1: Use ast.literal_eval to parse as a Python literal (e.g., list)
    try:
        evaluated_entry = ast.literal_eval(date_str_entry)
        if isinstance(evaluated_entry, list) and len(evaluated_entry) > 0:
            date_item = evaluated_entry[0]
            if isinstance(date_item, datetime.date):
                return pd.to_datetime(date_item.strftime('%Y-%m-%d'), errors='coerce')
            elif isinstance(date_item, str):
                return pd.to_datetime(date_item, errors='coerce')
        # If it's a non-list literal that somehow represents a date (e.g., just a datetime.date object not in a list string)
        elif isinstance(evaluated_entry, (datetime.date, str)):
             return pd.to_datetime(evaluated_entry, errors='coerce')

    except (ValueError, SyntaxError, TypeError):
        pass # literal_eval failed, fall through to regex attempts

    # Attempt 2: Use regex for patterns within the string (even if not a strict literal list or literal_eval failed)
    # Try to parse the format: datetime.date(YYYY, M, D) (without needing outer brackets to be matched by regex)
    match_dt_date = re.search(r'datetime\.date\((\d{4}),\s*(\d{1,2}),\s*(\d{1,2})\)', date_str_entry)
    if match_dt_date:
        try:
            year, month, day = map(int, match_dt_date.groups())
            return pd.to_datetime(f'{year}-{month}-{day}', errors='coerce')
        except ValueError:
            pass # Continue to next parsing attempt

    # Try to parse the format: 'YYYY-MM-DD' (without needing outer brackets to be matched by regex)
    match_iso_date = re.search(r"'(\\d{4}-\\d{2}-\\d{2})'", date_str_entry)
    if match_iso_date:
        date_iso_str = match_iso_date.group(1)
        return pd.to_datetime(date_iso_str, errors='coerce')

    # Attempt 3: If all else fails, try direct pd.to_datetime on the raw string itself
    return pd.to_datetime(date_str_entry, errors='coerce')


df['info.dates'] = df['info.dates'].apply(parse_date_entry_hybrid)

df['match_year'] = df['info.dates'].dt.year
df['match_month'] = df['info.dates'].dt.month
df['match_day'] = df['info.dates'].dt.day

print("Data types after CORRECTLY converting 'info.dates' and extracting components:\n")
df.info()
print("\nDataFrame after adding date components (first 5 rows):\n")
print(df.head())
```

```python
import ast
import datetime

# Restore 'info.dates' to its original string state, as it was corrupted to NaT in previous steps.
# This is done by re-reading only this specific column from the CSV.
df['info.dates'] = pd.read_csv('/content/International_T20_Data.csv')['info.dates']

def parse_info_dates_column(entry):
    if not isinstance(entry, str):
        return pd.NaT # Handle non-string entries (like NaT, though unlikely after restore)

    try:
        # Safely evaluate the string representation of a list
        date_list = ast.literal_eval(entry)
        if isinstance(date_list, list) and len(date_list) > 0:
            date_item = date_list[0]
            if isinstance(date_item, datetime.date):
                # Convert datetime.date object to string YYYY-MM-DD
                date_str = date_item.strftime('%Y-%m-%d')
            elif isinstance(date_item, str):
                # If it's already a string (e.g., 'YYYY-MM-DD'), use it directly
                date_str = date_item
            else:
                return pd.NaT

            return pd.to_datetime(date_str, errors='coerce')
        else:
            return pd.NaT # Empty list or not a list
    except (ValueError, SyntaxError, TypeError):
        # Fallback for entries that are not structured as list strings
        return pd.to_datetime(entry, errors='coerce') # Try direct conversion as last resort

df['info.dates'] = df['info.dates'].apply(parse_info_dates_column)

df['match_year'] = df['info.dates'].dt.year
df['match_month'] = df['info.dates'].dt.month
df['match_day'] = df['info.dates'].dt.day

print("Data types after CORRECTLY converting 'info.dates' and extracting components:\n")
df.info()
print("\nDataFrame after adding date components (first 5 rows):\n")
print(df.head())
```

```python
import re
import datetime

# Restore 'info.dates' to its original string state from the CSV
df['info.dates'] = pd.read_csv('/content/International_T20_Data.csv')['info.dates']

def parse_date_entry_robust(date_str_entry):
    if not isinstance(date_str_entry, str):
        return pd.NaT

    # Try to parse the format: [datetime.date(YYYY, M, D)]
    match_dt_date = re.search(r'datetime\.date\((\d{4}),\s*(\d{1,2}),\s*(\d{1,2})\)', date_str_entry)
    if match_dt_date:
        try:
            year, month, day = map(int, match_dt_date.groups())
            return pd.to_datetime(f'{year}-{month}-{day}', errors='coerce')
        except ValueError:
            pass # Continue to next parsing attempt

    # Try to parse the format: ['YYYY-MM-DD']
    match_iso_date = re.search(r"'(\d{4}-\d{2}-\d{2})'", date_str_entry)
    if match_iso_date:
        date_iso_str = match_iso_date.group(1)
        return pd.to_datetime(date_iso_str, errors='coerce')

    return pd.NaT # If no format matches

df['info.dates'] = df['info.dates'].apply(parse_date_entry_robust)

df['match_year'] = df['info.dates'].dt.year
df['match_month'] = df['info.dates'].dt.month
df['match_day'] = df['info.dates'].dt.day

print("Data types after CORRECTLY converting 'info.dates' and extracting components:\n")
df.info()
print("\nDataFrame after adding date components (first 5 rows):\n")
print(df.head())
```

```python
numerical_cols_to_fill_mean = ['info.outcome.by.wickets', 'Runs_margin']
categorical_cols_to_fill_mode = ['City', 'info.outcome.winner', 'info.gender']

for col in numerical_cols_to_fill_mean:
    if df[col].isnull().any():
        df[col] = df[col].fillna(df[col].mean())

for col in categorical_cols_to_fill_mode:
    if df[col].isnull().any():
        mode_val = df[col].mode()[0] # .mode() can return multiple if ties, so take the first
        df[col] = df[col].fillna(mode_val)

print("Missing values across all columns after handling:")
print(df.isnull().sum())
```
